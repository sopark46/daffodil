```python
import pandas as pd
import os
import datetime
import shutil
import matplotlib.pyplot as plt
```

```python
file_list = os.listdir("../all/")
file_list.sort()
dst_path = '../new_result/'
src_path = '../'
graphs_path = '../'
filename = ''
system_id = '29'
ess_id = '48'
```

```python
figsize_x = 16
figsize_y = 9
start_date = datetime.datetime(2016, 9, 1)
end_date = datetime.datetime(2018, 12, 31)
```

```python
def plot_all(df, column):
    import matplotlib.pyplot as plt
    
    fig = plt.figure(figsize=(figsize_x,figsize_y))#, tight_layout=True)
    fig.suptitle('ess_' + system_id + '_' + ess_id, fontsize=16)
    
    ax1 = fig.add_subplot(111)

    ax1.bar(df.index, 1, width=0.1, color='k', label='event')

#     ax1.set_ylim(bottom=79, top=101)
    ax1.set_xlim(left=start_date, right=end_date)

    ax1.legend(loc=2)
    plt.show()
    fig.savefig(graphs_path + 'ess_' + system_id + '_' + ess_id + ' event.png')
    plt.close()
#     ax1.set_title('ess_' + system_id + '_' + ess_id + ' SOH')
```

```python
def read_file(system_id, ess_id):
    global filename
    filename = 'ess_' + system_id + '_' + ess_id + '.csv'
    filename = 'event_history.csv'
    df = pd.read_csv(src_path + filename, parse_dates=['event_start_date',5,6], index_col = ['event_start_date'])
    return df
```

```python
df = read_file(system_id, ess_id)
```

```python
df.head()
```

```python
df1 = df.loc[df['ess_id'] == int(ess_id)]
```

```python
plot_all(df1, 'ess_id')
```
